```python
mfst_df = mfst_data[["Close"]].copy()
mfst_df  = mfst_df.reset_index().rename(columns={"index":"Date"})
mfst_df["Company"] = "Microsoft"
mfst_df
```

```python
ibm_df = ibm_data[["Close"]].copy()
ibm_df  = ibm_df.reset_index().rename(columns={"index":"Date"})
ibm_df["Company"] = "IBM"
ibm_df
```

```python
googl_df = googl_data[["Close"]].copy()
googl_df = googl_df.reset_index().rename(columns={"index":"Date"})
googl_df["Company"] = "Google"
googl_df
```

```python
nvdia_df = nvdia_data[["Close"]].copy()
nvdia_df = nvdia_df.reset_index().rename(columns={"index":"Date"})
nvdia_df["Company"] = "Nvidia"
nvdia_df
```

```python
msft_df = mfst_df[["Date","Close","Company"]]
ibm_df = ibm_df[["Date","Close","Company"]]
googl_df = googl_df[["Date","Close","Company"]]
nvdia_df = nvdia_df[["Date","Close","Company"]]
msft_df = msft_df.set_index("Date")
ibm_df = ibm_df.set_index("Date")
nvdia_df = nvdia_df.set_index("Date")
googl_df = googl_df.set_index("Date")
```

```python
msft_df = msft_df.rename(columns={"Close":"MSFT"})
ibm_df = ibm_df.rename(columns={"Close":"IBM"})
googl_df = googl_df.rename(columns={"Close":"GOOGL"})
nvdia_df = nvdia_df.rename(columns={"Close":"NVDA"})
```

```python
df = pd.concat([msft_df,ibm_df, googl_df, nvdia_df],axis = 1)
```

```python
df.head()
```

```python
df.info()
```

```python
new_cols = []
for col in df.columns:
    if isinstance(col, tuple):
        if col[0] == "Date":
            new_cols.append("Date")
        elif col[0] == "Company":
            new_cols.append("Company")
        elif col[0] == "Close":
            new_cols.append(col[1])
        else:
            new_cols.append("_".join([c for c in col if c]))
    else:
        new_cols.append(col)

df.columns = new_cols
```

```python
df.info()
```

```python
df.head()
```

```python
df = df.reset_index()
df.head()
```

```python
df.info()
```

```python
df_plot = df[["Date", "MSFT_MSFT", "IBM_IBM", "GOOGL_GOOGL", "NVDA_NVDA"]]
```

```python
df_plot = df_plot.rename(columns={"MSFT_MSFT":"MSFT","IBM_IBM":"IBM","GOOGL_GOOGL":"GOOGL","NVDA_NVDA":"NVDA"})
```

```python
df_plot.head()
```

```python
df_long = df_plot.melt(
    id_vars="Date",
    value_vars=["MSFT", "IBM", "GOOGL", "NVDA"],
    var_name="Company",
    value_name="Close"
)
```

```python
df_long
```

```python
sns.lineplot(data = df_long, x = "Date",y = "Close",hue = "Company")
plt.title("GOOGL/MSFT/IBM/NVDA Historical Stock Prices last 5 Years")
plt.xlabel("Date")
plt.ylabel("Closing Price $")
```
